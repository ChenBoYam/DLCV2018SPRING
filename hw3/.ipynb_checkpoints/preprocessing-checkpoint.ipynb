```python
import imageio
import os
import matplotlib.pyplot as plt
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader
%matplotlib inline
```

```python
dataset_path = "./hw3-train-validation/train/"
img = imageio.imread(os.path.join(dataset_path, "2311_sat.jpg"))
```

```python
# construct id list
image_id_list = list(set([item.split("_")[0] for item in os.listdir(dataset_path)]))
image_id_list[:10]
```

```python
print("image shape",img.shape)
```

```python
class SatelliteDataSet(Dataset):
    def encoder(pix):
        pix = list(pix)
        if pix == [0, 255, 255]:
            return 0
        elif pix == [255,255,0]:
            return 1
        elif pix == [255,0,255]:
            return 2
        elif pix == [0,255,0]:
            return 3
        elif pix == [0,0,255]:
            return 4
        elif pix == [255,255,255]:
            return 5
        else:
            return 6
        
    def __init__(self, train_or_valid = "train", transform=None):
        if train_or_valid == "train":
            dataset_path = "./hw3-train-validation/train/"
            self.image_id_list = list(set([item.split("_")[0] for item in os.listdir(dataset_path)]))
        elif train_or_valid == "valid":
            dataset_path = "./hw3-train-validation/validation/"
            self.image_id_list = list(set([item.split("_")[0] for item in os.listdir(dataset_path)]))
    
    def __getitem__(self, index):
        """
        Args:
            index: the index of item 
        Returns:
            image and its labels
        """
        X = imageio.imread(os.path.join(dataset_path, str(self.image_id_list[index])+"_sat.jpg"))
        y = imageio.imread(os.path.join(dataset_path, str(self.image_id_list[index])+"_mask.png"))
        # resize to (256,256,3) otherwise OOM occurs
        X = (np.array(X[::2,::2,])/255).transpose(2,0,1)
        y = np.array(y[::2,::2,])
        # onehot encoding for mask label
        y = np.array([encoder(pix) for pix in y.reshape(-1,3)]).reshape(256,256)
        return X,y
    def __len__(self):
        return len(self.image_id_list)
```

```python
train_dataset = SatelliteDataSet()
train_X = []
train_y = []
for i in range(len(train_dataset)):
    X, y = train_dataset[i]
    train_X.append(X)
    train_y.append(y)
```

```python
valid_dataset = SatelliteDataSet(train_or_valid="valid")
valid_X = []
valid_y = []
for i in range(len(valid_dataset)):
    X, y = valid_dataset[i]
    valid_X.append(X)
    valid_y.append(y)
```

```python
train_X = np.array(train_X)
train_y = np.array(train_y)
valid_X = np.array(valid_X)
valid_y = np.array(valid_y)
```

```python
np.save("train_X.npy", train_X)
np.save("train_y.npy", train_y)
np.save("valid_X.npy", valid_X)
np.save("valid_y.npy", valid_y)
```
